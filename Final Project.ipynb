### Handling Outliers <a id="handling-outliers"></a>

```python
# Iterate over each specified column to identify and remove outliers
def outliers_values(mental_health, columns = ['Technology_Usage_Hours', 'Social_Media_Usage_Hours',
                                              'Gaming_Hours', 'Screen_Time_Hours',
                                              'Physical_Activity_Hours']): 
    for column in columns:
         # Calculate the first quartile (25th percentile)
        Qr1 = mental_health[column].quantile(0.25)
        # Calculate the third quartile (75th percentile)
        Qr3 = mental_health[column].quantile(0.75)
        # Calculate the Interquartile Range (IQR)
        IQR = Qr3 - Qr1
       # Determine the upper value for outliers
        upper_extreme = Qr3 + 1.5 * IQR
        # Determine the lower value for outliers
        lower_extreme = Qr1 - 1.5 * IQR
       # Filter the DataFrame to exclude outliers
        mental_health = mental_health[(mental_health[column] >= lower_extreme) & (mental_health[column] <= upper_extreme)]
    return mental_health
 # Clean the mental_health DataFrame by removing outliers from specified columns   
mental_health_cleaned = outliers_values(mental_health)
```

#### Code explanation  
> First step: the function was defined with the data frame and a list of numeric columns that contain outliers to remove outliers<br>
> Second step: The for loop was made to iterate over each column in columns.<br> 
> Qr1 is the calculation of the first quartile in each column in columns.<br>
> Qr3 is the calculation of the third quartile in each column in columns.<br>
> IQR is inter quartile range  and it calculates the range between third quartile and first quartile.<br>
> Upper_extreme define the above outliers.<br>
> lower_extreme define the below outliers.<br>
> Third step filter out the data frame to exclude the outliers.<br>
> Fourth step return the cleaned data frame.<br>
> Final step function was called and stored in a new variable (mental_health_cleaned).

# Feature Engineering <a id="feature-engineering"></a>

#### Categorizing Age to groups <a id="categorize-age-to-groups"></a>

```python
# Function to categorize individuals into age groups based on their age
def categorize_age(age):
    if age >= 46:
        return 'Old Adult'   # Categorizes as 'Old Adult' for ages 46 and above
    elif age >= 31:
        return 'Adult'        # Categorizes as 'Adult' for ages 31 to 45   
    else:
        return 'Young Adult'   # Categorizes as 'Young Adult' for ages 30 and below        

# Applying the categorize_age function to the Age column and creating a new Age_Group column
mental_health_cleaned['Age_Group'] = mental_health_cleaned['Age'].apply(categorize_age)
```

#### Sleep hours categorization <a id="sleep-hours-categorization"></a>

```python
# Function to categorize sleep quality based on the number of sleep hours
def categorize_sleep_h(sleep):
    if sleep >= 6:
                # Return 'Well Sleeping' if sleep hours are 6 or more
        return 'Well Sleeping'
    elif sleep >= 4:
            # Return 'Fair Sleeping' if sleep hours are between 4 and 5.99
        return 'Fair Sleeping'
    else:
            # Return 'Bad Sleeping' if sleep hours are less than 4
        return 'Bad Sleeping'
# Applying the categorize_sleep_h function to the 'Sleep_Hours' column and creat new colum Named sleep_status
mental_health_cleaned['Sleep_Status'] = mental_health_cleaned['Sleep_Hours'].apply(categorize_sleep_h)
```
